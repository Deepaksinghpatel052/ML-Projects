```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df_feat = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])
df_feat.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, cancer['target'], test_size=0.33, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
prid = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, prid))
print("\n")
print(confusion_matrix(y_test, prid))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
perm_grid = {'C':[0.1,1,10,100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}
```

```python
grid_model = GridSearchCV(SVC(), perm_grid, verbose=1)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_pred = grid_model.predict(X_test)
```

```python
print(classification_report(y_test, grid_pred))
print("\n")
print(confusion_matrix(y_test, grid_pred))
```
