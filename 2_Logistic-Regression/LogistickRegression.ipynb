```python
def add_missing_value(col):
    Age = col[0]
    Pclass = col[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age  
        
```

```python
train_data['Age'] = train_data[['Age', 'Pclass']].apply(add_missing_value, axis=1)
```

```python
test_data['Age'] = test_data[['Age', 'Pclass']].apply(add_missing_value, axis=1)
```

```python
train_data.drop('Cabin',axis=1, inplace=True)
sns.heatmap(train_data.isnull(),yticklabels=False, cbar=False, cmap="viridis")
plt.show()
```

```python
train_data
```

```python
sex = pd.get_dummies(train_data['Sex'], drop_first=True)
Embard = pd.get_dummies(train_data['Embarked'])
```

```python
train_data = pd.concat([train_data, sex, Embard], axis=1)
```

```python
def get_ticket_no(ticket):
    return ticket.split(" ")[-1]


train_data['Ticket_No'] = train_data['Ticket'].apply(get_ticket_no)
train_data
```

```python
def get_name_data(name):
    return name.split(" ")[1]

train_data['Name_Set'] = train_data['Name'].apply(get_name_data)
#train_data.drop('Name_type',axis=1, inplace=True)
train_data
```

```python
Name_Set = pd.get_dummies(train_data['Name_Set'])

train_data = pd.concat([train_data, Name_Set], axis=1)
train_data
```

```python
train_data.drop(['PassengerId', 'Name', 'Name_Set', 'Sex', 'Ticket', 'Embarked'], axis=1, inplace=True)
train_data['Ticket_No'] = train_data['Ticket_No'].apply(lambda x: 0 if x=='LINE' else x )
train_data
```

```python
y = train_data['Survived']
X =  train_data.drop('Survived', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model = LogisticRegression(max_iter=10000)
lg_model.fit(X_train, y_train)
```

```python
pridict1 = lg_model.predict(X_test)
# pridict_test = lg_model.predict(test_data)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pridict1))
```

```python
print(confusion_matrix(y_test, pridict1))
```
